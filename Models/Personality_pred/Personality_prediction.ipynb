# **# Personality Prediction: Introvert or Extrovert**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
persona_data = pd.read_csv('/content/personality_dataset.csv') # accessing the data from storage using python pandas
```

```python
persona_data = persona_data.dropna()
```

```python
persona_data.replace({'Personality':{'Extrovert':0,'Introvert':1},'Stage_fear':{'No':0,'Yes':1},'Drained_after_socializing':{'No':0,'Yes':1}},inplace = True)


# replacing the values in the dataframe
```

```python
persona_data.head()
```

```python
x = persona_data.drop(columns = 'Personality', axis = 1) # devieding the features and labels x = all the features y = lables persnoality
y = persona_data['Personality']
```

# **Model training multiple**

```python
# list of models

models = [LogisticRegression(max_iter=1000),SVC(kernel='poly'),KNeighborsClassifier(),RandomForestClassifier()]
```

# **Cross validation**

Logistic Regression

```python
cv_score = cross_val_score(LogisticRegression(max_iter=1000),x,y,cv= 10) #cv = 5 means 5 fold  of the data
```

```python
print(round(cv_score.mean()*100,2))
```

creating a function for cross validation

```python
def compare_models_cv():
  for model in models:
    cv_score = cross_val_score(model,x,y,cv= 5)
    meanscore = round(cv_score.mean()*100,2)

    print('Cross Validation for this model',model, '=',cv_score)
    print('Accuracy score of the ', model, ' = ', meanscore)
    print('-'*50)
```

```python
compare_models_cv()
```

**the most accuracy is of Logistic regression and svc**

```python
best_models = {}
for name in models:
  grid = GridSearchCV(models[name],parameters[name], cv= 5)
  grid.fit(x,y)
  best_models[name] = grid.best_estimator_
  print(f"best parameters for {name} is {grid.best_params_}")
```

# **MAKING A PREDICTIVE SYSTEM**

```python
classifier = LogisticRegression(max_iter=1000)
```

training

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data = scaler.transform(x)
```

```python
x = standardized_data
y = persona_data['Personality']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y , test_size=0.2, stratify= y, random_state=2)
```

```python
classifier.fit(x_train,y_train)
```

Model evaluation

```python
x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

training data accuracy

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

test data accuracy

```python
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy score of the test data',test_data_accuracy)
```
